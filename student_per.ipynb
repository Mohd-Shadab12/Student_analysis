```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset (update the path to your dataset file location)
data=pd.read_csv("cleaned_student_data.csv")
# Display the first few rows to understand the dataset
print(data.head())
```

```python
# Get summary statistics for numeric columns
data.describe()
```

```python
numeric_data = data.select_dtypes(include=['float64', 'int64'])
# Calculate correlation matrix
correlation_matrix=numeric_data.corr()
# Create a heatmap to visualize the correlations
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm")
plt.title("Correlation_matrix")
plt.show()
```

```python
# Group by gender and calculate mean scores for each subject
gen_group=data.groupby('gender')[['math_score','history_score','physics_score','chemistry_score','biology_score','english_score','geography_score']].mean()
# Plot the results
gen_group.plot(kind='bar',figsize=(10,6))
plt.title('Average Subject scores by gender')
plt.ylabel('Average Score')
plt.show()
```

```python
# Group by career aspiration and calculate mean scores for each subject
carr_group=data.groupby('career_aspiration')[['math_score','history_score','physics_score','chemistry_score','biology_score','english_score','geography_score']].mean()
# Plot the results
carr_group.plot(kind='bar',figsize=(10,6))
plt.title('Average Subject scores by career aspiration')
plt.ylabel('Average Score')
plt.show()
```

```python
# Plot absence days vs average score in math
plt.figure(figsize=(8,3))
sns.scatterplot(x='absence_days',y='math_score',data=data)
plt.title('Absence days vs math score')
plt.xlabel('Absence Days')
plt.ylabel('Math score')
plt.show()
```

```python
# Plot absence days vs average score in biology
plt.figure(figsize=(8,3))
sns.scatterplot(x='absence_days',y='biology_score',data=data,color='brown')
plt.title('Absence days vs biology score')
plt.xlabel('Absence Days')
plt.ylabel('Biology score')
plt.show()
```

```python
# Boxplot for extracurricular activities vs study hours
plt.figure(figsize=(8, 3))
sns.boxplot(x='extracurricular_activities', y='weekly_self_study_hours', data=data,color='green')
plt.title('Extracurricular Activities vs Study Hours')
plt.xlabel('Extracurricular Activities')
plt.ylabel('Study Hours')
plt.show()
```

```python
# Boxplot for part-time job vs scores in math
plt.figure(figsize=(8, 3))
sns.boxplot(x='part_time_job', y='math_score', data=data)
plt.title('Part-Time Job vs Math Score')
plt.xlabel('Part-Time Job')
plt.ylabel('Math Score')
plt.show()
```

```python
# Boxplot for part-time job vs scores in chemistry
plt.figure(figsize=(8, 3))
sns.boxplot(x='part_time_job', y='chemistry_score', data=data,color='red')
plt.title('Part-Time Job vs Chemistry Score')
plt.xlabel('Part-Time Job')
plt.ylabel('Chemistry Score')
plt.show()
```

```python
# Create a heatmap to visualize the scores across subjects
plt.figure(figsize=(8, 3))
sns.heatmap(data[['math_score', 'physics_score', 'chemistry_score', 'biology_score', 'english_score']].corr(), annot=True, cmap="coolwarm")
plt.title('Subject Scores Correlation')
plt.show()
```

```python
# Create boxplot for each subject to detect outliers
plt.figure(figsize=(8, 3))
sns.boxplot(data=data[['math_score', 'physics_score', 'chemistry_score', 'biology_score', 'english_score']])
plt.title('Boxplot for Subject Scores (Outliers Detection)')
plt.show()
```
